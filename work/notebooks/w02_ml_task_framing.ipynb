## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

The pattern of content decline is too messy for a fixed `if-statement` due to several reasons:

1.  **Multi-dimensional Interaction**: Content performance is influenced by a multitude of factors (e.g., search volume, competition, content age, word count, CTR, engagement rate, average position, AI traffic percentage, and various 30-day and 90-day trend metrics). The 'declining' state (`trend_direction == 'down'`) is not just about one of these features but the complex, often non-linear, interactions between them.
2.  **Contextual Importance**: The significance of a metric's change can vary depending on other metrics. For example, a 10% drop in impressions might be critical for a high-volume page but negligible for a very low-volume page. A simple `if` rule struggles to apply context-sensitive thresholds.
3.  **Non-linear Relationships**: The relationship between features and content decline is rarely linear. A small change in one metric might have a large impact when combined with specific values in other metrics, while a large change might be irrelevant in another context. `If-statements` are inherently linear and additive in their decision-making.
4.  **Thresholding Challenges**: Defining fixed thresholds for each contributing factor (e.g., `trend_pct < -20%`, `avg_position > 15`) is arbitrary and prone to errors. What works for one content type or client might not work for another. Manually setting and maintaining these thresholds for all features and their combinations would be an enormous, if not impossible, task.
5.  **Dynamic Environment**: Search algorithms and user behavior constantly evolve. A fixed rule would quickly become outdated, requiring constant manual adjustments. Machine learning models, however, can adapt and learn new patterns from updated data.

In essence, machine learning models excel at discovering these intricate, hidden relationships and optimal decision boundaries across many features without requiring explicit, hand-coded rules, making them far more robust and accurate for identifying nuanced patterns like content decline.

```python
# Demonstrate the limitations of a simple 'if-statement' rule for identifying declining pages

print("Attempting to identify 'declining' pages with a simple rule:")

# Define a simple rule: e.g., trend_pct less than -20% AND impressions_90d greater than 100
simple_rule = (
    (df_filtered['trend_pct'] < -20) &
    (df_filtered['impressions_90d'] > 100)
)

# Actual declining pages
actual_declining = (df_filtered['trend_direction'] == 'down')

# Pages identified by the simple rule
identified_by_rule = df_filtered[simple_rule]

# Compare results
num_actual_declining = actual_declining.sum()
num_identified_by_rule = len(identified_by_rule)

# True Positives: Pages correctly identified by the rule as 'down'
true_positives = df_filtered[simple_rule & actual_declining]
num_true_positives = len(true_positives)

# False Negatives: Actual 'down' pages missed by the rule
false_negatives = df_filtered[~simple_rule & actual_declining]
num_false_negatives = len(false_negatives)

# False Positives: Pages incorrectly identified by the rule as 'down' (they are not actually 'down')
false_positives = df_filtered[simple_rule & ~actual_declining]
num_false_positives = len(false_positives)

print(f"\nTotal actual 'down' pages: {num_actual_declining:,}")
print(f"Total pages identified by simple rule: {num_identified_by_rule:,}")
print(f"  - Correctly identified (True Positives): {num_true_positives:,}")
print(f"  - Missed (False Negatives - actual 'down' but rule missed): {num_false_negatives:,}")
print(f"  - Incorrectly identified (False Positives - rule said 'down' but it's not): {num_false_positives:,}")

print("\nThis demonstrates that a simple fixed rule can lead to significant misses (false negatives) and incorrect flags (false positives), highlighting the need for a more sophisticated approach like ML.")
```
